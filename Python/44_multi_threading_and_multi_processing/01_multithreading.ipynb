## When to use Multi Threading

- **I/O bound tasks**: Tasks that spend more time waiting for I/O operations
- **When we want concurrent execution**

```python
import threading
import time
```

```python
def print_number():
    for i in range(5):
        print(f"Number:{i}")
        
t=time.time()
print_number()
finished_time=time.time()-t
print(finished_time)
```

```python
def print_letter():
    for l in "Thejas":
        print(f"Letter:{l}")

t=time.time()        
print_letter()
finished_time=time.time()-t
print(finished_time)
```

```python
def print_number():
    for i in range(5):
        print(f"Number:{i}")

def print_letter():
    for l in "Thejas":
        print(f"Letter:{l}")
        
t=time.time()
print_number()
print_letter()
finished_time=time.time()-t
print(finished_time)
```

```python
def print_number():
    for i in range(5):
        time.sleep(1)
        print(f"Number:{i}")

def print_letter():
    for l in "Thejas":
        time.sleep(1)
        print(f"Letter:{l}")
        
t=time.time()
print_number()
print_letter()
finished_time=time.time()-t
print(finished_time)
```

```python
def print_number():
    for i in range(5):
        time.sleep(1)
        print(f"Number:{i}")

def print_letter():
    for l in "Thejas":
        time.sleep(1)
        print(f"Letter:{l}")
        
# create 2 threads
t1=threading.Thread(target=print_number)
t2=threading.Thread(target=print_letter)

t=time.time()
# start the thread
t1.start()
t2.start()

# wait for the threads to complete
t1.join()
t2.join()
finished_time=time.time()-t
print(finished_time)
```
